# Busca informada
Usa conhecimento previo do problema para achar soluções mais eficientes que a busca cega.

## Best First Search
Usa uma função de avaliação aplicada em cada nó.

Se a função aplicada for uma estimativa heurística do objetivo, então temos uma busca gulosa, caso a função aplicada seja a profundidade do nó, temos uma busca em largura.

Podemos utilizar uma fila de prioridades, para ordenar os melhores resultados de f(n), sendo f a função e n cada nó.

### Busca Gulosa

Busca com função de avaliação igual a heurística, ou seja, uma estimativa de custo, para cada n, com relação ao nó final.

Um exemplo de função heuristica seria uma linha reta do ponto atual até o ponto final.

$$f(n) = h(n)$$

Por só considerar a heuristica do nó como função de avaliação, a busca gulosa possui algumas limitações:

* Não é ótima
* Não é completa
* Complexidade de tempo exponencial para o pior caso - $O(b^n)$, com b sendo a largura. Porém depende da heurística esolhida.
* Complexidade de espaço também exponencial - $O(b^n)$
