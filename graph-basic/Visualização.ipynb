# Vizualização de Grafos

```python
import networkx as nx
import matplotlib.pyplot as plt
```

### Grafo de Petersen

```python
G = nx.petersen_graph()
plt.subplot(121)
nx.draw(G, with_labels=True, font_weight='bold')
plt.subplot(122)
nx.draw_shell(
    G,
    nlist=[range(5, 10), range(5)],
    with_labels=True,
    font_weight='bold'
)
```

### Grafo planar completo

```python
K_5 = nx.complete_graph(5)

plt.plot()
nx.draw(
    K_5,
    with_labels=True,
    font_weight='bold'
)
```

### Grafo aleatório: modelo Erdõs-Rényi

```python
er = nx.erdos_renyi_graph(100, 0.15)
plt.figure(figsize=(15,15))
plt.plot()
nx.draw(
    er,
    with_labels=True,
    font_weight='bold'
)
```
