# Grafos - Conceitos Básicos
+ Grafos Direcionados (DiGrafos)
+ Grafos não-direcionados

## Grafo Direcionado
Grafo em que cada aresta tem uma direção, isto é, pode ser representada como <b>Tuplas (pares ordenados)</b> dos Vertices. 

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
DG = nx.DiGraph()
DG.add_nodes_from(
    ['v1', 'v2', 'v3', 'v4', 'v5']
)
DG.add_edges_from([('v1','v2'),
                 ('v1','v3'),
                 ('v2', 'v3'),
                 ('v3', 'v5'),
                 ('v4', 'v5')
                 ])
```

```python
plt.plot()
nx.draw(
    DG,
    with_labels=True,
    font_weight='bold'
)
```

## Grafos não direcionados
As relações representadas pelas arestas não tem sentido definido. As arestas podem ser seguidas em qualquer direção

```python
G =  nx.Graph(DG)
```

```python
nx.draw(
    G,
    with_labels=True,
    font_weight='bold'
)
```

Se $(u, v)$ é uma aresta no grafo, então dizemos que $v$ é <b>adjacente</b> a $u$.
+ Grafos direcionados: Se $u$ aponta para $v$, então o último é adjacente ao primeiro, porem, o contrário não é verdade.
+ Grafos não direcionados: temos a reciprocidade da adjacência. ou seja: $(u, v) \Leftrightarrow (v, u)$

### Caminhos no grafo
$$(v1, v2, v3)$$
$$(v3, v5, v4)$$
$$(v5, v4, v4)$$
Comprimento: Número de arestas percorridas no caminho.
$$compr(v1,v2,v3,v5) = 3$$

## Ciclos
+ Grafos dirigidos: formado por no mínimo, 1 aresta.
+ Grafos não dirigidos: formado, no mínimo, por 3 arestas.

### Grafo dirigido fortemente conexo
Existe um caminho entre qualquer par de vertices

```python
# Exemplo
DG2 = nx.DiGraph(DG)
DG2.add_edges_from([('v4', 'v2')])
DG2.remove_edges_from([('v1', 'v3')])
DG2.add_edges_from([('v3', 'v1')])

nx.draw(
    DG2,
    with_labels=True,
    font_weight='bold'
)
```

```python
DG3 = nx.DiGraph(DG)
DG3.clear()
DG3.add_weighted_edges_from([('v1', 'v2',-2)])
DG3.add_weighted_edges_from([('v1', 'v3', 12)])
DG3.add_weighted_edges_from([('v2', 'v3', 4)])
DG3.add_weighted_edges_from([('v3', 'v5', 40)])
DG3.add_weighted_edges_from([('v4', 'v5', 8)])
```

```python
w = nx.get_edge_attributes(DG2, 'weight')
nx.draw(
    DG2,
    with_labels=True,
    font_weight='bold'
)
```
